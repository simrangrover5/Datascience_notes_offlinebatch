### Mode
    
        1. More number of occurences 
        2. It is not always the mid value

```python
class Statistics:
    """This class will used to perform operations on numerical data"""
    def __init__(self, arr=None):
        self.arr = arr
    
    def mean(self):
        m = sum(self.arr)/len(self.arr)
        return m
    
    def median(self):
        N = len(self.arr)
        self.arr.sort()
        if N%2 == 0:
            i1 = N//2
            i2 = (N+2)//2
            m = (self.arr[i1-1] + self.arr[i2-1])/2
            return m
        else:
            i = (N+1)//2
            return self.arr[i-1]
    
    def mode(self):
        """Mode method will find the mode of array
        If we have 2 or 3 values with the same mode then it will 
        return the smallest value"""
        count = dict.fromkeys(self.arr, 0)
        # print(count)
        for val in self.arr:
            count[val] += 1
        max_freq = max(count.values())
        mode_ = []
        for key, value in count.items():
            if value == max_freq:
                mode_.append(key)
        return min(mode_)
```

```python
arr = [10, 20, 20, 21, 21, 21, 21, 22, 23, 24, 24, 24, 24, 25, 30]
s1 = Statistics(arr)
```

```python
s1.mode()
```

### Measures of Spread

    1. Deviation
    2. Standard Deviation
    3. Variance

#### Deviation

            dev = mean - x

```python
arr
```

```python
m = np.mean(arr)
dev = [i-m for i in arr]
```

```python
dev
```

```python
round(sum(dev)/len(arr))
```

    This tells us that the points are not getting deviate from mean 
    

```python
def deviation(arr):
    m = np.mean(arr)
    dev = [abs(i-m) for i in arr]
    return dev
```

```python
dev = deviation(arr)
avg_dev = sum(dev)/len(dev)
```

```python
avg_dev
```

```python
plt.figure(figsize=(10, 5))
plt.plot(arr, 'o')
plt.xticks(range(len(arr)))
m = np.mean(arr)
for i in range(len(arr)):
    plt.plot([i, i], [m, arr[i]], '--')
plt.plot([0, 19], [m, m], 'or--') 
plt.plot([0, 19], [m+avg_dev, m+avg_dev], 'ko--')
plt.plot([0, 19], [m-avg_dev, m-avg_dev], 'ko--')
plt.show()
```

### Drawback of deviation
        Deviation only covers the nearby values. So the large values
        get neclected
        
        So we will overcome this drawback using variance

### Variance

$$ variance = \frac{\sum\limits_{i=1}^N (x_i - \bar x)^2} {N} $$

```python
arr
```

```python
def variance(arr):
    dev = deviation(arr)
    var = sum([i**2 for i in dev])/len(arr)
    return var
```

```python
variance(arr)
```

```python
m
```

```python
plt.figure(figsize=(10, 5))
plt.plot(arr, 'o')
plt.xticks(range(len(arr)))
m = np.mean(arr)
var = variance(arr)
for i in range(len(arr)):
    plt.plot([i, i], [m, arr[i]], '--')
plt.plot([0, 19], [m, m], 'or--', label="MEAN") 
plt.plot([0, 19], [m+avg_dev, m+avg_dev], 'ko--', label='MEAN+DEV')
plt.plot([0, 19], [m-avg_dev, m-avg_dev], 'ko--', label="MEAN-DEV")
plt.plot([0, 19], [m+var, m+var], 'm--', label="MEAN+VAR")
plt.plot([0, 19], [m-var, m-var], 'm--', label="MEAN-VAR")
# plt.legend()
plt.show()
```

#### In Variance the outliers are also covered. Which will be solved by STD

```python
np.sqrt(var)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(arr, 'o')
plt.xticks(range(len(arr)))
m = np.mean(arr)
var = variance(arr)
std = np.sqrt(var)
for i in range(len(arr)):
    plt.plot([i, i], [m, arr[i]], '--')
plt.plot([0, 19], [m, m], 'or--', label="MEAN") 
plt.plot([0, 19], [m+avg_dev, m+avg_dev], 'ko--', label='MEAN+DEV')
plt.plot([0, 19], [m-avg_dev, m-avg_dev], 'ko--', label="MEAN-DEV")
plt.plot([0, 19], [m+var, m+var], 'm--', label="MEAN+VAR")
plt.plot([0, 19], [m-var, m-var], 'm--', label="MEAN-VAR")
plt.plot([0, 19], [m+std, m+std], 'g--', label="MEAN+VAR")
plt.plot([0, 19], [m-std, m-std], 'g--', label="MEAN-VAR")
# plt.legend()
plt.show()
```
