```python
url = "https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=mobiles&requestId=0dc08913-6136-4a6c-811f-1780635f5a6f&as-backfill=on"
```

```python
resp = requests.get(url)
flip = BeautifulSoup(resp.text, 'html')
```

```python
flip
```

```python
name = []
for n in flip.select('._4rR01T'):
    name.append(n.text)
```

```python
name
```

```python
price = []
for p in flip.findAll('div', attrs={'class': ['_1_WHN1']}):
    price.append(int(p.text[1:].replace(',', '')))
```

```python
price
```

```python
rating = []
for r in flip.select('._3LWZlK'):
    rating.append(eval(r.text))
```

```python
rating
```
