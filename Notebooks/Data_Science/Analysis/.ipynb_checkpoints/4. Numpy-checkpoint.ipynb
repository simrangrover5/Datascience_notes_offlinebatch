```python
deg = np.arange(0, 361, 15)
```

```python
sin = np.sin(deg)
cos = np.cos(deg)
plt.plot(deg, sin)
```

```python
rad = np.deg2rad(deg)
```

```python
sin = np.sin(rad)
cos = np.cos(rad)
plt.plot(rad, sin, 'r--')
plt.plot(rad, cos, 'b-.')
plt.show()
```

```python
tan = np.tan(rad)
plt.plot(rad, tan, 'm-.')
plt.show()
```

```python
sin = np.sin(rad)
cos = np.cos(rad)
plt.plot(rad, sin, 'r--')
plt.plot(rad, cos, 'b-.')
plt.plot(rad, tan, 'm:')
plt.show()
```

```python
img_copy = img.copy()
for i in range(len(img_copy)):
    for j in range(len(img_copy[i])):
        img_copy[i, j] = [sum(img_copy[i, j])//3]*3
```

```python
img_copy = img.copy()
for i in range(len(img_copy)//2):
    for j in range(len(img_copy[i])):
        img_copy[i, j] = [sum(img_copy[i, j])//3]*3
```

```python
plt.imshow(img_copy)
plt.axis('off')
plt.show()
```

```python
img_copy = img.copy()
for i in range(len(img_copy)//2):
    for j in range(len(img_copy[i])//2):
        img_copy[i, j] = [sum(img_copy[i, j])//3]*3
```

```python
plt.imshow(img_copy)
plt.axis('off')
plt.show()
```

```python
img_copy = img.copy()
for i in range(len(img_copy)//2):
    for j in range(len(img_copy[i])):
        img_copy[i, j] =[img_copy[i, j, 0]] + list(img_copy[i, j, [2, 1]])
```

```python
plt.imshow(img_copy)
plt.axis('off')
plt.show()
```
